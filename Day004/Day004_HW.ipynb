### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
import json
```

```python
# use 'r.text'
r = requests.get('https://api.github.com/events')
r.text
# type(r.text)
```

```python
type(json.loads(r.text))
```

```python
r.json
```

### 自行尋找一個合適的 API 接口做練習，並且查看其回傳內容

* https://cat-fact.herokuapp.com/facts (來源：https://alexwohlbruck.github.io/cat-facts/)
* http://odata.wra.gov.tw/v4/RealtimeWaterLevel (來源：https://data.gov.tw/dataset/25768)


```python
#使用get
cat_get = requests.get('https://cat-fact.herokuapp.com/facts')
json.loads(cat_get.text)
```

```python
#使用post
cat_post = requests.post('https://cat-fact.herokuapp.com/facts')
json.loads(cat_post.text)
```

```python
#使用get
odata_get = requests.get('http://odata.wra.gov.tw/v4/RealtimeWaterLevel')
json.loads(odata_get.text)
```

```python
#使用post
odata_post = requests.post('http://odata.wra.gov.tw/v4/RealtimeWaterLevel')
json.loads(odata_post.text)
```
