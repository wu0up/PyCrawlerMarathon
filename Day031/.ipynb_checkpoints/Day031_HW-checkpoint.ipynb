### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
header ={
    "user-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36",
    "authority": "zhihu-web-analytics.zhihu.com",
    "method": "POST",
    "path": "/api/v2/za/logs/batch",
    "scheme": "https",
    "accept": "*/*",
    "accept-encoding": "gzip, deflate, br",
    "accept-language": "zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6",
    "content-encoding": "gzip",
    "content-length": "370",
    "content-type": "application/x-protobuf",
    "origin": "https://www.zhihu.com",
    "referer": "https://www.zhihu.com/explore",
    "sec-fetch-mode": "cors",
    "sec-fetch-site": "same-site",
    "x-za-batch-size": '1',
    "x-za-clientid": "3d5ac3ac-6a8e-4033-b352-d95636aa3ffd",
    "x-za-log-version": "2.6.25",
    "x-za-platform": "DesktopWeb",
    "x-za-product": "Zhihu"
}

for key, value in header.items():
    data = requests.get('https://www.zhihu.com/explore', headers = {key:value})
    response = data.text
    if len(response)<180:
        print(key, ' is useless')
    else:
        print(key, ' is useful')
```
