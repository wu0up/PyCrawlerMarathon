### ① 取出今天所有的發文

```python
import re
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import pandas as pd
from itertools import groupby
import operator
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
article = requests.get(url)
soup = BeautifulSoup(article.text, "html5lib")
news= soup.findAll('h3')
article_list=[]
for i in news:
    if i.findAll('span', 'date'):
        tag = i.find('em').text
        href=i.find('a')['href']
        title = i.find('a').text
        article_list.append(
        {
            'title':title,
            'href':href,
            'tag':tag
        }
        )
article_list   
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
for key, groups in groupby(article_list, lambda d : d['tag']):
 # print(list(groups))
     #計算每個tag有幾個values
#     print('number of article: ', len(list(groups)))
    print(key)
    for item in groups:
        print(item)
    print('-----')
```

### ③ 哪一個類別的文章最多

```python
for key, groups in groupby(article_list, lambda d : d['tag']):
 # print(list(groups))
     #計算每個tag有幾個values
    count = len(list(groups))
    print('number of article: ', count)
    print(key)
    print('-----')
```
