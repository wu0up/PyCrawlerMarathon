### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
import xml.dom.minidom
import xmltodict
from urllib.request import urlretrieve, urlopen
import os, sys
from io import BytesIO
from zipfile import ZipFile
import shutil 
from pathlib import Path
import json
import xml.etree.ElementTree as ET
import glob
```

```python
#將xml檔案解壓後，存到Data資料夾
path = 'http://pycrawler.cupoy.com/file-download/part03/example.zip'
file_path = './Data'

with urlopen(path) as zipresp:
    with ZipFile(BytesIO(zipresp.read())) as zfile:
        zfile.extractall(file_path)
        filenames = zfile.namelist()
print(filenames)
```

```python
try:
    with open('./Data/Homework.xml', "w") as outfile: 
        for fname in filenames: 
            with open('./Data/'+fname) as infile:
                outfile.write(infile.readline())
except:
    print('發生錯誤！')  #希望可以print error code
```

```python
#使用I/O 讀取檔案

with open('./Data/09007_Week24_CH.xml', 'r') as fh:
    f = fh.read()
print(f)   #print(f) vs print(fh)
```

```python
#使用xmltodict讀取檔案
with open('./Data/09007_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))
print(doc)
```

```python
#use xml.dom開啟檔案 ->why 沒印出東西
import xml.dom.minidom as xmldom

# 得到文件物件
domobj = xmldom.parse('./Data/09007_72hr_CH.xml')
print("xmldom.parse:", type(domobj))
# 得到元素物件
elementobj = domobj.documentElement
print ("domobj.documentElement:", type(elementobj))
#獲得子標籤
subElementObj = elementobj.getElementsByTagName("login")
print ("getElementsByTagName:", type(subElementObj))
print (len(subElementObj))
# 獲得標籤屬性值
print (subElementObj[0].getAttribute("username"))
print (subElementObj[0].getAttribute("passwd"))
#區分相同標籤名的標籤
subElementObj1 = elementobj.getElementsByTagName("caption")
for i in range(len(subElementObj1)):
    print ("subElementObj1[i]:", type(subElementObj1[i]))
    print (subElementObj1[i].firstChild.data) #顯示標籤對之間的資料
```

```python
#使用xml.etree進行檔案讀取
tree = ET.ElementTree('./Data/09007_72hr_CH.xml')
root = tree.getroot()
for elem in tree.iter():
    print(elem.tag, elem.attrib, elem.text)
```

I/O的方式：印出保留XML格式的string

xmltodict的方式：形成字典的形式

python 解析XML檔案的方式：

|  名稱  | 說明  | 
|  ----  | ----  |
| SAX (simple API for XML )  | Python 标准库包含 SAX 解析器，SAX 用事件驱动模型，通过在解析XML的过程中触发一个个的事件并调用用户定义的回调函数来处理XML文件。 | 
| DOM(Document Object Model)  | 将 XML 数据在内存中解析成一个树，通过对树的操作来操作XML；比較慢，耗內存 | 
| ElementTree | ElementTree就像一个轻量级的DOM，具有方便友好的API。代码可用性好，速度快，消耗内存少。 | 

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
with open('./Data/64_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))
# print(doc)

#data1 vs data
data1 = doc['cwbopendata']['dataset']['locations']['location']

for i in range(len(data1)):
    data=doc['cwbopendata']['dataset']['locations']['location'][i]['locationName']
    print(data)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for i in range(len(data1)):
    loc =doc['cwbopendata']['dataset']['locations']['location'][i]['locationName'] 
    date=doc['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][0]['dataTime']
    temp=doc['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][0]['elementValue']['value']
    degree=doc['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][0]['elementValue']['measures']
    print(loc ,date, temp, degree)
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
data_1=doc['cwbopendata']['dataset']['locations']['location'][0]['weatherElement']
for i in range(len(data_1)):
    loc =doc['cwbopendata']['dataset']['locations']['location'][0]['locationName'] 
    date=doc['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time'][i]['dataTime']
    temp=doc['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][i]['elementValue']['value']
    degree=doc['cwbopendata']['dataset']['locations']['location'][i]['weatherElement'][0]['time'][i]['elementValue']['measures']
    print(loc ,date, temp, degree)
```
