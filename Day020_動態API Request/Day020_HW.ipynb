```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import re
data_dict ={}
table_lst = table.find_all('td')
for ele in table_lst:
    if "style" in ele.attrs:
        name = ele.text
        data_dict[name]={}
    else:
        if re.compile('\d{4}\/\d{2}').match(ele.text):
            date = ele.text
            data_dict[name][date]={}
        elif re.compile('\d+').match(ele.text):
            value = ele.text
            data_dict[name][date]=value
```

```python
import pandas as pd
pd.DataFrame(data_dict)
```
